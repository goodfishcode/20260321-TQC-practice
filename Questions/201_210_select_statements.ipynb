來源：[doc/TQC_Python_201_210_select_statements.md](../doc/TQC_Python_201_210_select_statements.md)

使用方式：
- 每題請實作 `solve_題號(input_data: str) -> str`
- 完成後直接執行該題下方的測試 cell
- 測試比對會保留空白與換行差異，只忽略最後多出的空白行


````python
def normalize_output(text):
    if not isinstance(text, str):
        text = str(text)
    text = text.replace("\r\n", "\n")
    return text.rstrip("\n")


def run_example_tests(problem_code, func, cases):
    if not cases:
        print(f"{problem_code}: 找不到可自動解析的完整範例，請自行補測試。")
        return

    for idx, case in enumerate(cases, start=1):
        actual = normalize_output(func(case["input"]))
        expected = normalize_output(case["expected"])
        if actual != expected:
            raise AssertionError(
                f"[{problem_code} case {idx}]\n"
                f"input:\n{case['input']!r}\n\n"
                f"expected:\n{expected!r}\n\n"
                f"actual:\n{actual!r}"
            )

    print(f"{problem_code}: {len(cases)} 個範例測試通過。")
````

## 204 算術運算
題目說明：請開啟 PYD204.py 檔案，依題意進行作答，依輸入值進行算術運算；作答完成請另存新檔為 PYA204.py 再進行評分。  
設計說明：輸入兩個整數 a、b，然後再輸入一個算術運算子（+、-、*、/、//、%），輸出運算結果。  
輸入說明：兩個整數 a、b，及一個算術運算子  
輸出說明：運算結果（無須做格式化）  

範例輸入：
```
30
20
*
```
範例輸出：
```
600
```

---

````python
# 老師解答
a = int(input())
b = int(input())
operator = input()

if operator == '+':
    print(a + b)
elif operator == '-':
    print(a - b)
elif operator == '*':
    print(a * b)
elif operator == '/':
    print(a / b)
elif operator == '//':
    print(a // b)
else:
    print(a % b)
````

````python
def solve_204(input_data: str) -> str:
    """讀入整份標準輸入字串，回傳整份標準輸出字串。"""
    a_input, b_input, operator = input_data.split()
    a = int(a_input)
    b = int(b_input)

    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        return a / b
    elif operator == '//':
        return a // b
    else:
        return a % b

    raise NotImplementedError("Implement solve_204")
````

````python
cases_204 = [
{
    'input': '30\n20\n*',
    'expected': '600',
}
]
run_example_tests("204", solve_204, cases_204)
````

````python
# 自訂測試區
# custom_input = """"""
# print(solve_204(custom_input))
a = int(input())
b = int(input())
c = input()

if c == '+':
    result = a + b
elif c == '-':
    result = a - b
elif c == '*':
    result = a * b
elif c == '/':
    result = a / b
elif c == '//':
    result = a // b
else:
    result = a % b

print(result)
````
