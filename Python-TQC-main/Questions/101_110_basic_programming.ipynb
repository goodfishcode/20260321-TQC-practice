# TQC+ 程式語言Python 第1類：基本程式設計（101–110 題目整理）

來源：[doc/TQC_Python_101_110_basic_programming.md](../doc/TQC_Python_101_110_basic_programming.md)

使用方式：
- 每題請實作 `solve_題號(input_data: str) -> str`
- 完成後直接執行該題下方的測試 cell
- 測試比對會保留空白與換行差異，只忽略最後多出的空白行


```python
def normalize_output(text):
    if not isinstance(text, str):
        text = str(text)
    text = text.replace("\r\n", "\n")
    return text.rstrip("\n")


def run_example_tests(problem_code, func, cases):
    if not cases:
        print(f"{problem_code}: 找不到可自動解析的完整範例，請自行補測試。")
        return

    for idx, case in enumerate(cases, start=1):
        actual = normalize_output(func(case["input"]))
        expected = normalize_output(case["expected"])
        if actual != expected:
            raise AssertionError(
                f"[{problem_code} case {idx}]\n"
                f"input:\n{case['input']!r}\n\n"
                f"expected:\n{expected!r}\n\n"
                f"actual:\n{actual!r}"
            )

    print(f"{problem_code}: {len(cases)} 個範例測試通過。")
```

## 101 整數格式化輸出

題目說明：請開啟 PYD101.py 檔案，依題意進行作答，進行整數的格式化輸出；作答完成請另存新檔為 PYA101.py 再進行評分。  
設計說明：輸入四個整數，輸出格式要求：每列兩個整數、寬度皆為 5、間隔 1 個空白；前兩列靠右對齊、後兩列靠左對齊，左右以直線（`|`）為邊界。  
輸入說明：四個整數  
輸出說明：四個整數的格式化輸出（每列兩個）  
範例輸入：
85  
4  
299  
478  
範例輸出：
| 85 4|  
| 299 478|  
|85 4 |  
|299 478 |

```python
def solve_101(input_data: str) -> str:
    """讀入整份標準輸入字串，回傳整份標準輸出字串。"""
    values = [int(line.strip()) for line in input_data.splitlines() if line.strip()]
    a, b, c, d = values
    lines = [
        f"| {a} {b}|",
        f"| {c} {d}|",
        f"|{a} {b} |",
        f"|{c} {d} |",
    ]
    return "\n".join(lines)
```

```python
cases_101 = [
{
    'input': '85\n4\n299\n478',
    'expected': '| 85 4|\n| 299 478|\n|85 4 |\n|299 478 |',
}
]
run_example_tests("101", solve_101, cases_101)
```
